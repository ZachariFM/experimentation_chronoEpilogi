```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_p = pd.read_csv("./results/SynthBase_params.csv",compression="gzip")
df_r = pd.read_csv("./results/SynthBase_results.csv",compression="gzip")
df_all = pd.merge(df_p, df_r, on=["trial","experiment_timestamp"])
```

```python
u = df_all.groupby(["fs_name","forecaster_name","data_name"])[['fs_size']].mean().reset_index()
print("number of MTS evaluated by fs and es algo")
u.groupby(["fs_name","forecaster_name"])["fs_size"].count()
```

```python
print("Average performances over configuration and MTS, during tuning, for each fs and es algo")
df_all.groupby(["fs_name","forecaster_name"])[['fs_size', 'eq-tp',
       'eq-fp', 'eq-fn', 'eq-precision', 'eq-recall', 'eq-f1', 'fs_time',
       'forecaster.training_time', 'forecaster.predict_time', 'mfe', 'mae',
       'rmse', 'r2', 'mape', 'smape', 'mase']].mean()
```

```python
print("Average performances over MTS, for the best configuration each, during tuning, for each fs and es algo")
idx = df_all.groupby(["fs_name","forecaster_name","data_name"])['r2'].transform(max) == df_all['r2']
u=df_all[idx].groupby(["fs_name","data_name","forecaster_name"]).first().reset_index()
u.groupby(["fs_name","forecaster_name"])[['fs_size', 'eq-tp',
       'eq-fp', 'eq-fn', 'eq-precision', 'eq-recall', 'eq-f1', 'fs_time',
       'forecaster.training_time', 'forecaster.predict_time', 'mfe', 'mae',
       'rmse', 'r2', 'mape', 'smape', 'mase']].mean()
```

```python
print("Performance, For each MTS, for the best configuration each, during tuning, for each fs and es algo")
idx = df_all.groupby(["fs_name","forecaster_name","data_name"])['r2'].transform(max) == df_all['r2']
df_all[idx].groupby(["data_name","fs_name","forecaster_name"]).first()[['fs_size', 'eq-tp',
       'eq-fp', 'eq-fn', 'eq-precision', 'eq-recall', 'eq-f1', 'fs_time',
       'forecaster.training_time', 'forecaster.predict_time', 'mfe', 'mae',
       'rmse', 'r2', 'mape', 'smape', 'mase']]
```
